```python
import cv2
from matplotlib import pyplot as plt
#Import image
image = cv2.imread(r"ch_0/lena.png") # this is read in BGR format

plt.imshow(image)
plt.show()
```

```python
rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)# this converts it into RGB
plt.imshow(rgb_img)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
print(gray.shape)
```

```python
# converting the image from rgb to gray
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
img_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv)
```

```python
img_yuv = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)
plt.imshow(img_yuv)
```
