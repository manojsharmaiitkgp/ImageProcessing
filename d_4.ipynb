rotate and translate

```python
from PIL import Image
import matplotlib.pyplot as plt

# Read image
fn="ch_0/lena.png"
img=Image.open(fn)
```

```python
def myDisplay(img,y):
    plt.subplot(1,2,1),plt.imshow(img)
    plt.subplot(1,2,2),plt.imshow(y)
```

```python
# display original and rotated image
angle=45
rotated_img=img.rotate(angle)
myDisplay(img,rotated_img)
```

```python
# clock wise rotation
# angle=-45
angle=-90
rotated_img=img.rotate(angle)
# display original and rotated image
myDisplay(img,rotated_img)
```

```python
print(img.size)
```

```python
#  Resize image to new dimenstion
newsize=(200,200)
# Resized image
resize_img=img.resize(newsize)
# display resized image
myDisplay(img,resize_img)
```

```python
newImg = img.rotate(45, resample=Image.BICUBIC)
myDisplay(img,newImg)
```

```python
import cv2
import numpy as np

# Read image
img = cv2.imread(fn) 
plt.imshow(img)
```

```python
# The number of pixels
num_rows, num_cols = img.shape[:2]
x1=70
y1=110

# Creating a translation matrix
translation_matrix = np.float32([ [1,0,x1], [0,1,y1] ])

# Image translation
img_translation = cv2.warpAffine(img, translation_matrix, (num_cols,num_rows))
```

```python
plt.imshow(img_translation)
```
