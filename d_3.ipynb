splitting and merging the image

```python
from PIL import Image
import matplotlib.pyplot as plt


fn="ch_0/lena.png"
# open the original image
img=Image.open(fn)
plt.imshow(img)
```

```python
# here img is PIL image
print(type(img))
```

```python
import numpy as np
#  convert PIL to numpy array
img_np=np.array(img)
```

```python
print(type(img_np))
plt.imshow(img_np)
print(img_np.shape)
```

```python
rimg=img_np[:,:,0]
gimg=img_np[:,:,1]
bimg=img_np[:,:,2]
```

```python
plt.figure(figsize=(8,8))
plt.subplot(1,3,1), plt.imshow(rimg)
plt.subplot(1,3,2), plt.imshow(gimg)
plt.subplot(1,3,3), plt.imshow(bimg)
```

```python
plt.figure(figsize=(8,8))
plt.subplot(1,3,1), plt.imshow(rimg,cmap='gray')
plt.subplot(1,3,2), plt.imshow(gimg,cmap='gray')
plt.subplot(1,3,3), plt.imshow(bimg,cmap='gray')
```

```python
plt.figure(figsize=(3, 3))
plt.subplot(1,3,1), plt.imshow(rimg,cmap='gray')
plt.subplot(1,3,2), plt.imshow(gimg,cmap='gray')
plt.subplot(1,3,3), plt.imshow(bimg,cmap='gray')
```

```python
plt.figure(figsize=(12, 12))
plt.subplot(1,3,1), plt.imshow(rimg,cmap='gray')
plt.subplot(1,3,2), plt.imshow(gimg,cmap='gray')
plt.subplot(1,3,3), plt.imshow(bimg,cmap='gray')
```

```python
number_of_channel=len(img.split())
print(number_of_channel)
r, g, b = img.split()
```

```python
plt.imshow(r,cmap='gray')
```

```python
im1 = Image.merge( 'RGB', (r, g, b))
```

```python
plt.imshow(im1)
```
