```python
import matplotlib.pyplot as plt

data = [1,2,2,3,3,3,4,4,5]

plt.hist(data)

plt.title('How to plot a simple histogram in matplotlib ?', fontsize=10)

plt.savefig("plot_simple_histogramme_matplotlib_01.png")

plt.show()
```

```python
plt.hist(data, normed=1)

plt.title('How to plot a simple histogram in matplotlib ?', fontsize=10)

plt.savefig("plot_simple_histogramme_matplotlib_02.png")

plt.show()
```

```python
plt.hist(data, bins = 2)

plt.title('How to plot a simple histogram in matplotlib ?', fontsize=10)

plt.savefig("plot_simple_histogramme_matplotlib_03.png")

plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

data = np.random.randn(1000)
```

```python
plt.hist(data);
```

```python
plt.hist(data, bins=30, normed=True, alpha=0.5,
         histtype='stepfilled', color='steelblue',
         edgecolor='none');
```

```python
plt.rcParams['figure.figsize'] = 10, 4
data = np.zeros((1000))
l = plt.hist(data,normed = True, bins = 100)
plt.axvline(0.01, lw=1)
plt.ylim(0, 150)
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```
