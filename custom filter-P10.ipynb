```python
from PIL import ImageFilter, Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# kernel =ImageFilter.Kernel((3,3),[1,0,-1,5,0,-5,1,0,1])
kernel =ImageFilter.Kernel((3,3),[1,0,-1,5,0,-5,-1,0,-1])
# kernel =ImageFilter.Kernel((3,3),[-1,0,1,-2,10,2,-1,0,1])
```

```python
fn='ch_02/images/image.jpg'
img=Image.open(fn)
```

```python
plt.imshow(img)
```

```python
img=img.convert("L")
# img1=np.array(img)/255
n_img=img.filter(kernel)
```

```python
def myDisplayCmap(x,y):
    plt.subplot(1,2,1),plt.imshow(x,cmap='gray')
    plt.subplot(1,2,2),plt.imshow(y,cmap='gray')
```

```python
def myDisplay(x,y):
    plt.subplot(1,2,1),plt.imshow(x)
    plt.subplot(1,2,2),plt.imshow(y)
```

```python
plt.figure(figsize=(15,15))
myDisplayCmap(img,n_img)
```

```python
from skimage import filters

edge=filters.sobel(img)

print('min(img)=',np.min(img))
print('min(edge)=',np.min(edge))
print('min(n_img)=',np.min(n_img))


print('max(img)=',np.max(img))
print('max(edge)=',np.max(edge))
print('max(n_img)=',np.max(n_img))
```

```python
plt.figure(figsize=(15,15))
myDisplayCmap(img,edge)
```
