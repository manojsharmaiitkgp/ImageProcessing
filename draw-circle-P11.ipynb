```python
import numpy as np
from skimage import io, draw
```

```python
img=np.zeros((100,100),dtype=np.uint8)
top=40
right=70
radius=10
x,y=draw.circle(top,right,radius)
img[x,y]=1
io.imshow(img,cmap='gray')
io.show()
```

```python
import numpy as np
from skimage import io, draw
img=np.zeros((100,100),dtype=np.uint8)
a=20
b=40
c=10
d=40
x,y=draw.ellipse(a,b,c,d)
img[x,y]=1
io.imshow(img)
io.show()
```

```python
import numpy as np
from skimage import io,draw
img=np.zeros((100,100),dtype=np.uint8)
r=np.array([10,25,80,50])
c=np.array([10,60,40,10])
x,y=draw.polygon(r,c)
img[x,y]=1
io.imshow(img)
io.show()
```
