Approach1:

```python
from PIL import Image
import matplotlib.pyplot as plt
# Read image 
fn="ch_0/lena.png"
img=Image.open(fn)
```

```python
# get image dimensions  
img.size
```

```python
print(type(img))
```

```python
# get pixel value at given paint
point=(200,300)
img.getpixel(point)
```

```python
# display image
plt.imshow(img)
```

```python
# img1=img.convert("L")
# gray scale conversion
img1=img.convert("LA") 

#get pixel value
value1=img1.getpixel(point) 
print(value1)
plt.imshow(img1)
```

Approach2:

```python
#reading image using imageio library
import imageio
import matplotlib.pyplot as plt
# read image file
fn="ch_0/lena.png"
im = imageio.imread(fn)
# this returns the number of channel along with height and width
print('im.shape=',im.shape)
# print the type of im
print('type(im)=',type(im))
# display image
plt.imshow(im)
```

Applocah 3:

```python
# This block of code read the image in numpy format no explict conversion is required
from skimage import io
import matplotlib.pyplot as plt
# https://www.tutorialspoint.com/how-can-scikit-learn-be-used-to-convert-an-image-from-rgb-to-grayscale-in-python
fn="ch_0/lena.png"
im = io.imread(fn)
# this returns the number of channel along with height and width
print('im.shape=',im.shape)
# print the type of im
print('type(im)=',type(im))
# display image
plt.imshow(im)
```

```python
# Conversion to gray scale
from skimage.color import rgb2gray
grayscale_img = rgb2gray(im)
print(type(grayscale_img))

plt.imshow(grayscale_img,cmap='gray')
```
