Reading and croping

```python
from PIL import Image
import matplotlib.pyplot as plt
# specify the dimention to crop (top left, bottom right cordinate)
dim=(100,100,400,400)
fn="ch_0/lena.png"
# open the original image
img=Image.open(fn)
# crop the image using given dimention
crop_image=img.crop(dim)
# displaying the original image
plt.imshow(img)
```

```python
# displaying the crop image
plt.imshow(crop_image)
```

Displaying image using subplot

```python
# to display the image in subplot
plt.subplot(1,2,1),plt.imshow(img)
plt.subplot(1,2,2),plt.imshow(crop_image)
```

saving image

```python
# output_fn="lena_crop.png"
output_fn="lena_crop_1.jpeg"
crop_image.save(output_fn)
```
