Image Enhancement

```python
from PIL import Image
import matplotlib.pyplot as plt

# Read image
fn="ch_0/lena.png"
img=Image.open(fn)
```

```python
def myDisplay(x,y):
    plt.subplot(1,2,1),plt.imshow(x)
    plt.subplot(1,2,2),plt.imshow(y)
```

```python
from PIL import ImageEnhance
```

```python
img_br=ImageEnhance.Brightness(img)
print(type(img_br))
```

```python
#  We need to pass the enhance parameter also
# else it will be error
myDisplay(img,img_br.enhance(15))
```

```python
#  We need to pass the enhance parameter also
# else it will be error
myDisplay(img,img_br.enhance(5))
```

```python
vec=[15,5,2,1.6, 1.2, 1, 0.8, 0.5, 0.2, 0 , -0.5,-1 ]
cnt=len(vec)
print(cnt)
```

```python
i=0
plt.figure(figsize=(15,15))
for obj in vec:
    i=i+1
    plt.subplot(3,4,i), plt.imshow(img_br.enhance(obj))
    plt.subplots_adjust(wspace=None, hspace=None)
    plt.title("obj="+str(obj))
```

```python
# https://stackoverflow.com/questions/20057260/how-to-remove-gaps-between-subplots-in-matplotlib/45487279
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

plt.figure(figsize = (15,15))
gs1 = gridspec.GridSpec(2, 6)
gs1.update(wspace=0.5, hspace=0.00) # set the spacing between axes. 


i=-1
plt.figure(figsize=(15,15))
for obj in vec:
    i=i+1

    ax1 = plt.subplot(gs1[i])
    plt.imshow(img_br.enhance(obj))
    plt.title("obj="+str(obj))
```
